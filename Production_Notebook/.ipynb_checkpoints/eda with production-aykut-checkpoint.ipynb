## EDA with clean dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_eda = pd.read_csv('df_cleaned.csv',low_memory=False)
```

```python
df_edanum = df_eda.select_dtypes(include=np.number) # Getting numerical data 
```

```python



vol_cols = [col for col in df_edanum.columns if "volume" in col]
span_cols =[col for col in df_edanum.columns if "spannung" in col]
piezo_cols = [col for col in df_edanum.columns if "piezo" in col]
oeo_cols=[col for col in df_edanum.columns if "oeo" in col]
ns1_cols = [col for col in df_edanum.columns if "ns1" in col if 'geraet' not in col]
ns2_cols = [col for col in df_edanum.columns if "ns2" in col if 'geraet' not in col]
pulse_col = [col for col in df_edanum.columns if "puls" in col]

columns_to_use = vol_cols + span_cols + piezo_cols + oeo_cols + ns1_cols + ns2_cols + pulse_col


df_to_use = df_edanum[columns_to_use]
```

```python
df_to_use
```

```python
# Plotting boxplots of ns2 columns to decide if NaN values should be replaced with mean or median



fig, axes = plt.subplots(3,2, figsize=(10,20))
sns.boxplot(ax=axes[0,0], x=df_to_use.ns2_center_variation)
sns.boxplot(ax=axes[0,1], x=df_to_use.ns2_front_angle)
sns.boxplot(ax=axes[1,0], x=df_to_use.ns2_curve)
sns.boxplot(ax=axes[1,1], x=df_to_use.ns2_15xlength)
sns.boxplot(ax=axes[2,0], x=df_to_use.ns2_parabolic_factor)
```
