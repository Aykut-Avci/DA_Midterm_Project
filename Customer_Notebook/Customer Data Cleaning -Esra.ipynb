### Create A Dictionary for Unique Customer Ids

```python
#sorted the customer name to give them index customer id
customer_list_sorted = sorted(customer_list)
keys = customer_list_sorted
```

```python
#customer id list for the values of dict
values =list(range(1,len(customer_list_sorted)+1))
values
```

We created the keys and values of our dictionary now with the function we'll assign them to each other.

```python
# create dictionary with our customer name and id to assign them each other.
def create_dictionary(keys, values):
    result = {} # empty dictionary
    for key, value in zip(keys, values):
        result[key] = value
    return result
# use create_dictionary func.
print(create_dictionary(keys=keys, values=values))

create_dict= create_dictionary(keys=keys, values=values)
```

```python
#create a dictionary for each unique customer
create_dict
```

After creating the dictionary now we'll use the values for our new column which called customer_id.

```python
#Assign the values from dictionary to a new column 

new_customer_anon_id = []

for customer in customer_df["customer"]:
    new_customer_anon_id.append(create_dict[customer])
    
new_customer_anon_id

customer_df.insert(1, "customer_id", new_customer_anon_id)
```
